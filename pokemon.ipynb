```python
class Pokemon:
    def __init__(self, name, level, typo, max_health, health, is_knocked_out):
        self.name = name
        self.level = level
        self.typo = typo
        self.max_health = max_health
        self.health = health
        self.is_knocked_out = is_knocked_out
        
        
    def attack(self, other_pokemon):
        damage = 0
        if(self.typo == other_pokemon.typo or self.typo == "fire" and other_pokemon.typo == "water"):
            damage = self.level * 1.5
            print('The Attack has sorted effect')
            other_pokemon.lose_health(damage)
        elif ((self.typo == "electric" and other_pokemon.typo == "water") or (self.typo == "water" and other_pokemon.typo == "fire")):
            damage = self.level *  2
            print('The Attack has sorted effect')
            other_pokemon.lose_health(damage)
        else:
            print('The attack had no effect on the opponent') 
      
    def lose_health(self, damage):
        self.health = self.health - damage
        if (self.health <= 0):
            self.knocked_out()
        else:
            print('{} has {} of energy.'.format(self.name, self.health))

        
    def gain_health(self):
        recover_power = self.health * 2
        self.health = self.health + recover_power
        if (self.health > self.max_health):
            return '{} has recovered now it has {} of energy!'.format(self.name, self.max_health)
        else:
            return '{} has recovered now it has {} of energy!'.format(self.name, self.health)
    
    def knocked_out(self):
        self.is_knocked_out = 'yes'
        print('{} is dead.'.format(self.name))
        
    def revive_pokemon(self):
        self.health = self.max_health
        self.is_knocked_out = 'no'
        print ('{} is now fit again!'.format(self.name))
        

        
    
```

```python
class Trainer:
    def __init__(self, name, potions):  
        self.name = name
        self.potions = potions
        self.pokemons = []
        self.current_pokemon = None
    
    def __repr__(self):
        return 'New trainer has been created, his name is {}'.format(self.name)
    
    def add_pokemon(self, pokemon):
        if type(pokemon) is Pokemon:
            if pokemon not in self.pokemons:
                if len(self.pokemons) < 6:
                    self.pokemons.append(pokemon)
                    self.current_pokemon = self.pokemons[0]
                    print('{} added!'.format(pokemon.name))
                else:
                    print('No space for more pokemons!')
            else:
                print('You already have this pokemon!')
    
    def switch_pokemon(self):
        if len(self.pokemons) > 1:
            for pokemon in self.pokemons:
                if pokemon != self.current_pokemon:
                    self.current_pokemon = pokemon
                    break
                else:
                    continue
            print('{} is now active!'.format(self.current_pokemon.name))
        else:
            print('You only have one Pokemon!')
    
    def activate_potion(self):
        if (self.current_pokemon.health == self.current_pokemon.max_health):
            print('Your pokemon does not need to be revived')
        else:
            if (self.potions != 0):
                self.current_pokemon.revive_pokemon()
                self.potions -= 1
                print(self.potions)
            else:
                print('You have no potions left!')
    
    def activate_attack(self, other_trainer):
        self.current_pokemon.attack(other_trainer.current_pokemon)
            
   
    
```

```python
pokemon1 = Pokemon('pikatchu', 4, 'electric', 25, 20, 'no')
pokemon2 = Pokemon('charizard', 5,'fire', 20, 15, 'no')
pokemon3 = Pokemon('Ho-Oh', 5, 'fire', 20, 15, 'no')
trainer1 = Trainer('Joao', 2)
trainer2 = Trainer('Miguel', 3)

trainer1.add_pokemon(pokemon1)
trainer1.add_pokemon(pokemon2)
trainer1.switch_pokemon()
trainer2.add_pokemon(pokemon3)

trainer1.activate_attack(trainer2)







```
